```python
from graphstats import *
import numpy as np
import matplotlib.pyplot as plt
from graspy import *
from graspy.plot import pairplot
```

```python
n = 200
seed_ratio = 0.1

# Class 1 prior
pi1 = 0.5

# Class 2 prior
pi2 = 1 - pi1
pis = np.array([pi1, pi2])

P = np.array([[0.6, 0.3], [0.3, 0.45]])
A = adj_matrix(n, pis, P, acorn = 2)

plt.imshow(A)
```

```python
ase = embed.AdjacencySpectralEmbed(n_elbows = 2)
Xhat = ase.fit_transform(A)
pairplot(Xhat)
```

```python
GMM = cluster.GaussianCluster()
hmm = GMM.fit(Xhat)
dir(hmm.model_)
```
